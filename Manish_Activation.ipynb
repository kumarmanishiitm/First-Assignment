```python
import numpy as np

class Sigmoid():
    def __init__(self, c=1, b=0):
        self.c = c
        self.b = b

    def value_of_x(self, x):
        val = 1 + np.exp(-self.c*(x + self.b))
        return 1/val

    def differentiation(self, x, remove=False):
        y = self.value(x)
        if remove==True:
            y = y[:-1,:]
        val = self.c*y*(1-y)
        return val
class Softmax():
    def __init__(self):
        pass

    def value_of_x(self, x):
        val = np.exp(x)/np.sum(np.exp(x), axis=0)
        return val

    def differentiation(self, x):
        y = self.value(x)
        mat = np.tile(y, y.shape[0])
        val = np.diag(y.reshape(-1,)) - (mat*mat.T)
        return val

class Relu():
    def __init__(self):
        pass

    def value_of_x(self, x):
        val = x
        val[val<0] = 0
        return val

    def differentiation(self, x):
        val = np.ones(x.shape)
        val[val<=0] = 0
        return val

class Tanh():
    def __init__(self):
        pass

    def value_of_x(self, x):
        num = np.exp(x) - np.exp(-x)
        denom = np.exp(x) + np.exp(-x)
        return num/denom

    def differentiation(self, x):
        y = self.value(x)
        val = 1 - y**2
        return val
```
